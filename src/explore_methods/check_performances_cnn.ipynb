```python
import pandas as pd
import json
import os
```

```python
path_data = "../../results/raw1/cnn_explorer/"

list_epochs_explorer = os.listdir(path_data)
```

```python
matrix_data = []

for epoch in list_epochs_explorer:
    list_docs = os.listdir("{}{}".format(path_data, epoch))
    for doc in list_docs:
        with open("{}{}/{}".format(path_data, epoch, doc)) as doc_open:
            doc_info = json.load(doc_open)

            name_doc = doc.replace("coding_", "").split(".json")[0][:-2]

            r2_testing = None
            try:
                r2_testing = doc_info["test_metrics"]["r2"]
            except:
                r2_testing = doc_info["test_metrics"]["r2_score"]

            row_data = [
                name_doc,
                doc_info["arquitecture"],
                doc_info["trainable_params"],
                doc_info["non_trainable_params"],
                doc_info["train_metrics"]["mse"],
                doc_info["train_metrics"]["mae"],
                doc_info["train_metrics"]["r2_score"],
                doc_info["train_metrics"]["pearsonr"],
                doc_info["train_metrics"]["spearmanr"],
                doc_info["test_metrics"]["mse"],
                doc_info["test_metrics"]["mae"],
                r2_testing,
                doc_info["test_metrics"]["pearsonr"],
                doc_info["test_metrics"]["spearmanr"], 
                doc_info["total_time"],
                doc_info["epochs"]
            ]
            matrix_data.append(row_data)

df_summary = pd.DataFrame(
    matrix_data, 
    columns=['coding', 'architecture', 'trainable_params', 'non_trainable_params', 'mse_train', 'mae_train', 'r2_score_train', 'pearsonr_train', 'spearmanr_train', 'mse_test', 'mae_test', 'r2_score_test', 'pearsonr_test', 'spearmanr_test', 'total_time', 'epochs'])
```

```python
df_summary = df_summary.dropna()
```

```python
df_summary = df_summary.sort_values(by=['r2_score_test'], ascending=False)
df_summary
```
